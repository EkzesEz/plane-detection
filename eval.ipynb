## Making CSV file

```python
amount_of_frames = 242
```

```python
import pandas as pd
```

```python
coords = {"x": [], "y": []}

for i in range(1, amount_of_frames + 1):
    try:
        with open(f"C:/Workspace/YOLO/runs/detect/predict7/labels/test_{i}.txt") as f:
            data = f.readline().split()
            coords["x"].append(data[1])
            coords["y"].append(data[2])
    except:
        continue
```

```python
print(coords)
```

```python
pd.DataFrame(coords).to_csv("coords.csv", index=False)
```
